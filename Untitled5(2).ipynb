```python
# Import Library Dasar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import Library Machine Learning (Scikit-Learn)
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Setup visualisasi
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")
```

1. **DATA UNDERSTANDING**

```python
# 1. Load Dataset (Pastikan file berada di folder yang sama atau sesuaikan path-nya)
df = pd.read_csv('FraudShield_Banking_Data.csv')

# Menampilkan 5 baris pertama
display(df.head())

# Menampilkan info dataset (tipe data dan missing values)
print("\n--- Informasi Dataset ---")
df.info()

# Statistik deskriptif data numerik
display(df.describe())

# Mengecek keseimbangan kelas target (Fraud vs Normal)
print("\n--- Distribusi Target (Fraud_Label) ---")
print(df['Fraud_Label'].value_counts())
```

2. **DATA PREPROCESSING**

```python
# 1. Menghapus kolom yang tidak memiliki nilai prediktif (ID, IP, Waktu yang rumit)
columns_to_drop = ['Transaction_ID', 'Customer_ID', 'Device_ID', 'IP_Address', 'Transaction_Time', 'Transaction_Date']
df_clean = df.drop(columns=columns_to_drop, errors='ignore')

# 2. Mengatasi Missing Values (Drop baris yang kosong untuk kesederhanaan, atau isi dengan median/modus)
df_clean = df_clean.dropna()

# 3. Encoding Data Kategorikal
# a. Ubah Target (Fraud_Label) jadi angka: Normal = 0, Fraud = 1
df_clean['Fraud_Label'] = df_clean['Fraud_Label'].map({'Normal': 0, 'Fraud': 1})

# b. Ubah kolom Yes/No menjadi 1/0
yes_no_cols = ['Is_International_Transaction', 'Is_New_Merchant', 'Unusual_Time_Transaction']
for col in yes_no_cols:
    df_clean[col] = df_clean[col].map({'Yes': 1, 'No': 0})

# c. One-Hot Encoding untuk kolom kategorikal lainnya (Tipe transaksi, lokasi, dll)
categorical_cols = ['Transaction_Type', 'Merchant_Category', 'Transaction_Location', 'Customer_Home_Location', 'Card_Type']
df_clean = pd.get_dummies(df_clean, columns=categorical_cols, drop_first=True)

# 4. Memisahkan Fitur (X) dan Target (y)
X = df_clean.drop('Fraud_Label', axis=1)
y = df_clean['Fraud_Label']

# 5. Feature Scaling (Standarisasi agar algoritma KNN bekerja maksimal)
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print("Shape data setelah preprocessing:", X_scaled.shape)
```

3. **Exploratory Data Analysis (EDA)**

```python
plt.figure(figsize=(14, 5))

# Visualisasi 1: Distribusi Target
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Fraud_Label', palette='Set2')
plt.title('Distribusi Kelas Target (Normal vs Fraud)')

# Visualisasi 2: Hubungan Amount vs Fraud
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Fraud_Label', y='Transaction_Amount (in Million)', palette='Set1')
plt.title('Distribusi Nominal Transaksi Terhadap Fraud')

plt.tight_layout()
plt.show()
```

4. **Data Splitting**

```python
# Skenario 1: 70% Train, 30% Test
X_train_70, X_test_70, y_train_70, y_test_70 = train_test_split(X_scaled, y, test_size=0.30, random_state=42, stratify=y)

# Skenario 2: 80% Train, 20% Test
X_train_80, X_test_80, y_train_80, y_test_80 = train_test_split(X_scaled, y, test_size=0.20, random_state=42, stratify=y)

# Skenario 3: 90% Train, 10% Test
X_train_90, X_test_90, y_train_90, y_test_90 = train_test_split(X_scaled, y, test_size=0.10, random_state=42, stratify=y)

# Simpan dalam dictionary agar mudah di-loop nanti
splits = {
    "70:30": (X_train_70, X_test_70, y_train_70, y_test_70),
    "80:20": (X_train_80, X_test_80, y_train_80, y_test_80),
    "90:10": (X_train_90, X_test_90, y_train_90, y_test_90)
}

print("Pembagian data berhasil dilakukan!")
```

5, 6 & 7: Model Building, Hyperparameter Tuning & Evaluation

```python
# Siapkan list untuk menyimpan hasil evaluasi
results = []

# Definisikan Algoritma dan Parameter Grid untuk Tuning
models = {
    "Decision Tree": {
        "model": DecisionTreeClassifier(random_state=42),
        "params": {'max_depth': [5, 10, 15, None], 'min_samples_split': [2, 5, 10]}
    },
    "KNN": {
        "model": KNeighborsClassifier(),
        "params": {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
    },
    "Random Forest": {
        "model": RandomForestClassifier(random_state=42),
        "params": {'n_estimators': [50, 100], 'max_depth': [10, 20, None]}
    }
}

# Loop untuk setiap Skenario Splitting
for split_name, (X_train, X_test, y_train, y_test) in splits.items():
    print(f"\n================ MENGGUNAKAN DATA SPLIT {split_name} ================")

    # Loop untuk setiap Model
    for model_name, mp in models.items():
        print(f"Training & Tuning {model_name}...")

        # Grid Search / Hyperparameter Tuning
        clf = GridSearchCV(mp['model'], mp['params'], cv=3, scoring='accuracy', n_jobs=-1)
        clf.fit(X_train, y_train)

        # Prediksi menggunakan model terbaik (best parameters)
        best_model = clf.best_estimator_
        y_pred = best_model.predict(X_test)

        # Evaluasi
        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, zero_division=0)
        rec = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)

        # Simpan ke dalam list
        results.append({
            "Model": model_name,
            "Data Split": split_name,
            "Best Params": str(clf.best_params_),
            "Accuracy": acc,
            "Precision": prec,
            "Recall": rec,
            "F1-Score": f1
        })

        # Tambahkan ini di bawah perhitungan F1-Score pada Tahap 5,6,7
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(4, 3))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.title(f'Confusion Matrix: {model_name} ({split_name})')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.show()

        print(f"--> {model_name} | Akurasi: {acc:.4f} | Best Params: {clf.best_params_}")

print("\nSeluruh proses training selesai!")
```

**8. Model Comparison**

```python
# 1. Membuat Tabel Perbandingan
df_results = pd.DataFrame(results)
display(df_results)

# 2. Membuat Grafik Perbandingan Akurasi (Bar Chart)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_results, x="Data Split", y="Accuracy", hue="Model", palette="viridis")

# Menambahkan garis target minimal (85% dan 95%)
plt.axhline(y=0.95, color='r', linestyle='--', label='Target 95%')
plt.axhline(y=0.85, color='orange', linestyle='--', label='Target 85%')

plt.title('Perbandingan Akurasi Model pada Berbagai Skenario Data Splitting')
plt.ylim(0, 1.1)
plt.ylabel('Accuracy')
plt.xlabel('Rasio Data Splitting')
plt.legend(loc='lower right')
plt.show()
```

**KESIMPULAN**

Algoritma Terbaik: Random Forest terbukti menjadi algoritma yang paling tangguh dan akurat untuk mendeteksi anomali pada data tabular perbankan ini.

Pengaruh Rasio Data: Perubahan rasio pembagian data (70:30, 80:20, 90:10) memberikan dampak pada hasil akurasi, di mana semakin banyak data training umumnya model semakin mengenali pola, namun rentan overfitting.
